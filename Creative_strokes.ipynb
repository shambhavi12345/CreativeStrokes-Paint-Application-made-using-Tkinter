```python
#CREATIVE STROKES


from tkinter import *
import tkinter.font
from tkinter.ttk import Scale
from tkinter import colorchooser, filedialog, messagebox
import PIL.ImageGrab as ImageGrab
from time import strftime


class Paint:

    def __init__(self,r):
        
        #layout of the canvas
        self.r = r
        self.r.title("CREATIVE STROKES")
        self.r.geometry('1000x520')
        self.r.configure(background="white")
        self.r.resizable(True,True)
        self.pen_color = "black"
        self.eraser_color = "white"
        self.r.iconbitmap("paint.ico")

        #layout of the color window
        self.color_frame = LabelFrame(self.r,text='Colors',font=('Times New Roman',13,'bold'),bd=5,relief = RIDGE,bg="white")
        self.color_frame.place(x=0,y=0,width=70,height=185)


        #list of colors
        colors = ['#FFFFCC','#00FFFF','#FFFF00','#FFFFFF','#000000','#00CC00','#FF66FF','#000033','#Fe771E','#FE57A1','#1AB1AB',
                 '#FE110E','#10ADED']
        i=j=0
        for color in colors:
            Button(self.color_frame,bg=color,bd=2,relief=RIDGE,width=3,command= lambda col = color:self.select_color(col)).grid(row=i,column=j)
            i+=1
            if i==6:
                i=0
                j=1

        #Buttons
        self.eraser_button = Button(self.r, text="CANVAS", bd=4, bg='white',background='gold',command=self.canavas_color, width=6, relief=RIDGE)
        self.eraser_button.place(x=0, y=190)
        self.eraser_button = Button(self.r,text="ERASER",bd=4,bg='white',background='gold',activebackground='orange red',command=self.eraser,width=6,relief=RIDGE)
        self.eraser_button.place(x=0,y=217)
        self.eraser_button = Button(self.r, text="SAVE", bd=4, bg='white',background='SpringGreen2', command=self.save_art,width=6, relief=RIDGE)
        self.eraser_button.place(x=0, y=267)
        self.eraser_button = Button(self.r, text="CLEAR", bd=4, bg='white',background='Orange red',command=lambda: self.canvas.delete("all"), width=6, relief=RIDGE)
        self.eraser_button.place(x=0, y=297)
        
        
        #Scale bar
        self.size_scale_frame = LabelFrame(self.r,text='Size',bd=5,bg='turquoise',font=('Times New Roman',15,'bold'),relief=RIDGE)
        self.size_scale_frame.place(x=0,y=345,height=170,width=70)

        self.size_scale = Scale(self.size_scale_frame,orient=VERTICAL,from_ =50,to=0,length=120)
        self.size_scale.set(1)
        self.size_scale.grid(row=0,column=1,padx=15)
        
        #Canvas
        self.canvas = Canvas(self.r,bg='white',bd=5,relief=GROOVE,height=500,width=7000)
        self.canvas.place(x=80,y=0)

        #Mouse drag
        self.canvas.bind("<B1-Motion>",self.paint)
       
    
    
    #Commands for buttons
    def paint(self,event):
        x1,y1 = (event.x-2),(event.y-2)
        x2,y2 = (event.x+2),(event.y+2)

        self.canvas.create_oval(x1,y1,x2,y2,fill=self.pen_color,outline=self.pen_color,width=self.size_scale.get())

    def select_color(self,col):
        self.pen_color=col

    def eraser(self):
        self.pen_color= self.eraser_color

    def canavas_color(self):
        color = colorchooser.askcolor()
        print(color)
        self.canvas.configure(background=color[1])
        self.eraser_color = color[1]

    def save_art(self):
        try:
            filename = filedialog.asksaveasfilename(defaultextension='.jpg')
            print(filename)
            x = self.r.winfo_rx()+self.canvas.winfo_x()
            print(x, self.canvas.winfo_x())
            y = self.r.winfo_rooty() + self.canvas.winfo_y()
            print(y)
            x1 = x + self.canvas.winfo_width()
            print(x1)
            y1 = y + self.canvas.winfo_height()
            print(y1)
            ImageGrab.grab().crop((x,y,x1,y1)).save(filename)
            messagebox.showinfo('Paint says','You image is saved as'+str(filename))
        except:
            pass

        
        
if __name__ == "__main__":
    r = Tk()
    p = Paint(r)
    r.mainloop()
```
